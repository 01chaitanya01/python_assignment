```python
import numpy as np
import pandas as pd

column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data", header= None, delimiter="\s+", names=column_names)

dataFrame = pd.DataFrame(data)

dataFrame.head()
```

```python
dataFrame.describe()
```

```python
x = dataFrame["RM"].values.reshape(-1,1)
y = dataFrame["MEDV"].values.reshape(-1,1)

print(f'X : {x[:5]}')
print(f'Y : {y[:5]}')
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x_train, y_train)
```

```python
y_test_predict = model.predict(y_test)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, y_test_predict)
```

```python
import matplotlib.pyplot as plt

y_train_predict = model.predict(x_train)

plt.scatter(x_train, y_train)
plt.plot(x_train, y_train_predict, color="red")
plt.xlabel("actual value")
plt.ylabel("predited value")
```
